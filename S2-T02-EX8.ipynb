```python
import numpy as np
import matplotlib.pyplot as plt #Importa llibrería matplot
```

```python
import matplotlib.image as mpimg
```

```python
image = mpimg.imread('TestRGB.png')
plt.imshow(image)
```

```python
print(image)
```

```python
imageGray = np.mean(image,-1)
plt.imshow(imageGray)
```

```python
plt.imshow(image*0.5)#Ajuste de luminosidad
```

```python
#Detecta Vermells
plt.imshow(image[:,:,0], vmin=0,vmax=1)
plt.title("Vermell")
```

```python
#Detecta Verds
plt.imshow(image[:,:,1], vmin=0,vmax=1)
plt.title("Verd")
```

```python
#Detecta Blaus
plt.imshow(image[:,:,2], vmin=0,vmax=1)
plt.title("Blau")
```

```python
#copies imatges
image_RedBlue = image.copy()
image_RedGreen = image.copy()
image_GreenBlue = image.copy()
```

```python
#filtra color verd
image_RedBlue[:,:,1] = 0
plt.title("Imatge filtrats el verd")
plt.imshow(image_RedBlue)
```

```python
#filtra color blau
image_RedGreen[:,:,2] = 0
plt.title("Imatge filtrats el blau")
plt.imshow(image_RedGreen)
```

```python
#filtra color vermell
image_GreenBlue[:,:,0] = 0
plt.title("Imatge filtrats el vermell")
plt.imshow(image_GreenBlue)
```

```python
#copia imatjes
plt.imsave("imatgeFiltreVermell.png",image_GreenBlue)
```

```python
plt.imsave("imatgeFiltreBlau.jpg",image_RedGreen)
plt.imsave("imatgeFiltreVerd.png",image_RedBlue)
```
